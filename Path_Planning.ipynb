```python
#import labraries
import numpy as np
```

```python
#defining the environment
environment_rows = 11
environment_coloumns = 11
#intilizing q learning
q_values= np.zeros ((environment_rows, environment_coloumns, 4))



```

```python
#defining actions
actions = [ 'up','right','down','left']
```

```python
#rewards
rewards = np.full((environment_rows,environment_coloumns), -100)
rewards[0,5] = 100; #reward for goal state
#defining aisles locations(white squares)
aisles = {} # store location
aisles[1] = [i for i in range (1,10)]
aisles[2] = [1, 7, 9]
aisles[3] = [i for i in range (1,8)]
aisles[3].append(9)
aisles[4] = [3,7]
aisles[5] = [i for i in range(11)]
aisles[6] = [5]
aisles[7] = [i for i in range (1,10)]
aisles[8] = [3,7]
aisles[9] = [i for i in range (11)]

#set "-1" rewards for all aisles equations
for row_index in range(1,10):
    for coloumn_index in aisles[row_index]:
      rewards[row_index, coloumn_index] = -1.


```

```python
#define a fuction that determines terminal state
def is_terminal_state(current_row_index,current_coloumn_index):
  if rewards[current_row_index,current_coloumn_index] == -1:
    return False
  else:
    return True

#define a function that will choose a random , non - terminal location
def get_starting_location():
  #getting random row and coloumn index
  current_row_index = np.random.randint(environment_rows)
  current_coloumn_index = np.random.randint(environment_coloumns)
  while is_terminal_state(current_row_index,current_coloumn_index):
    current_row_index = np.random.randint(environment_rows)
    current_coloumn_index = np.random.randint(environment_coloumns)
  return current_row_index, current_coloumn_index

#define episilon greedy algorithm that will choose next action
def get_next_action(current_row_index, current_coloumn_index, epsilon):
  if np.random.random() < epsilon :
    return np.argmax(q_values[current_row_index, current_coloumn_index])
  else:
    return np.random.randint(4)

#define a function that willl get the next location based on choosen action
def get_next_location(current_row_index, current_coloumn_index, action_index):
  new_row_index = current_row_index
  new_coloumn_index = current_row_index
  if actions[action_index] == 'up' and current_row_index > 0:
    new_row_index -= 1
  elif actions[action_index] == 'right' and current_coloumn_index < environment_coloumns - 1:
    new_coloumn_index += 1
  elif actions[action_index] == 'down' and current_row_index < environment_rows - 1:
    new_row_index += 1
  elif actions[action_index] == 'left' and current_coloumn_index < 0 :
    new_coloumn_index -= 1
  return new_row_index, new_coloumn_index

#define a fuction to get shortest path from any location within the ware house
def get_shortest_path(start_row_index, start_coloumn_index):
  if is_terminal_state(start_row_index, start_coloumn_index):
    return []
  else :
    current_row_index, current_coloumn_index = start_row_index, start_coloumn_index
    shortest_path = []
    shortest_path.append([current_row_index, current_coloumn_index])
    while not is_terminal_state(current_row_index,current_coloumn_index):
      action_index = get_next_action(current_row_index,current_coloumn_index, 1)
      current_row_index, current_coloumn_index = get_next_location(current_row_index, current_coloumn_index, action_index)
      shortest_path.append([current_row_index, current_coloumn_index])
    return shortest_path




```

```python
##Training##
#defining parameters
epsilon = 0.9
discount_factor = 0.9
learning_rate = 0.9


for episode in range(1000) :

  row_index, coloumn_index = get_starting_location()


  while not is_terminal_state(row_index, coloumn_index):

    action_index = get_next_action(row_index, coloumn_index, epsilon)

    old_row_index, old_coloumn_index = row_index, coloumn_index

    row_index, coloumn_index = get_next_location(row_index, coloumn_index, action_index)



    reward = rewards[row_index, coloumn_index]

    old_q_value = q_values[old_row_index, old_coloumn_index, action_index]



    temporal_difference = reward + (discount_factor * np.max(q_values[row_index,coloumn_index])) - old_q_value

    new_q_value = old_q_value + (learning_rate * temporal_difference)

    q_values[old_row_index, old_coloumn_index, action_index] = new_q_value


print("TRAINING  COMPLETE !!")
print(q_values)



```
